```python
#Importar archivo csv, separador '|'
df3 = pd.read_excel('/content/raw-customer-churn-data.xlsx')
#Hechar un vistaso a los primeros registros
df3.tail(20)
```

```python
#Eliminar espacios al comienzo y al final
df3 = df3.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df3.head(20)
```

```python
#Remplazamos los campos vacios por 'No register'
#df3['Churn'] = df3['Churn'].fillna('No register')
```

```python
#Las personas que no registraon genero las cambiamos por 'Otro', debido a que cada vez mas gente tiene problemas con eso
#df3['Gender'] = df3['Gender'].fillna('Otro')
```

```python
#Las personas que no registraron su genero no se tienen en cuenta
#df3['Gender'] =df3['Gender'].dropna()
df3 = df3.dropna(subset=['Gender'])

df3.head(10)
```

```python
#Hay personas que m quieren decir male y f quieren decir female
df3['Gender'] = df3['Gender'].replace('m', 'male')
df3['Gender'] = df3['Gender'].replace('f', 'female')
```

```python
#Para conservar la mayor cantidad de informacion, ponemos la edad del registro anterior a la persona que no puso edad
df3['Age']=df3['Age'].fillna(method='bfill')
```

```python
df3 = df3.drop_duplicates(subset=['Name'])
```
