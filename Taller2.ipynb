```python
#Cargamos los datos y mostramos algunos de ellos
drinks = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
drinks.head()
#dentro de head(10) podemos insertar el numeor de registro que queremos ver
#drinks.tail() similarmente podemos ver los ultimos registros de la tabla
```

```python
#Agrupamos por continente y sacamos la medie de la cervez que toman
drinks.groupby('continent').beer_servings.mean()
#EU es el contenente que mas bebe
#AS es el contienente que menos bebe
```

```python
#Para cada continente imprime las estadÃ­sticas de consumo de vino.
drinks.groupby('continent').wine_servings.describe()
```

```python
#Imprime el consumo medio de alcohol por continente para cada columna
drinks.groupby('continent').mean()
```

```python
#Imprime la mediana del consumo de alcohol por continente para cada columna
drinks.groupby('continent').median()
```

```python
#Imprime los valores medio, mÃ­nimo y mÃ¡ximo para el consumo de bebidas
#espirituosas.
#Esta vez genera un DataFrame
drinks.groupby('continent').spirit_servings.agg(['mean','min','max'])
```
