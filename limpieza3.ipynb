```python
maxi=np.max(movies_df['vote_average'])
mini=np.min(movies_df['vote_average'])
```

```python
record = movies_df[movies_df['vote_average'] == maxi]
record.head()
```

```python
def weighted_rating(x,m=m,c=c):
  v=x['vote_count']
  R=x['vote_average']
  return (v/(v+m)*R)+(m/(m+v)*c)
```

```python
q_movies = movies_df.copy().loc[movies_df['vote_count'] >= m]
```

```python
q_movies['Score']= q_movies.apply(weighted_rating,axis=1)
```

```python
q_movies = q_movies.sort_values('Score', ascending=False)
q_movies[['title','vote_count', 'vote_average', 'Score']].head(15)
```

```python
q_movies.shape
```

```python
q_movies['overview'].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
q_movies['overview'] = q_movies['overview'].fillna('')
tfidf_matrix = tfidf.fit_transform(q_movies['overview'])
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
consine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
consine_sim.shape
```

```python
indices = pd.Series(q_movies.index, index = q_movies['title']).drop_duplicates()
```

```python
def recomendar(title, consine_sim = consine_sim):
  idx = indices[title]
  sim_scores = list(enumerate(consine_sim[idx]))
  sim_scores= sorted(sim_scores, key=lambda x: x[1], reverse = True)
  sim_scores = sim_scores[1:11]
  movies_indices = [i[0] for i in sim_scores]
  return q_movies['title'].iloc[movies_indices]
```

```python
recomendar('Forrest Gump')
```

```python
def is_integer(n):
    try:
        float_n = float(n)
        int_n = int(float_n)
    except ValueError:
        return False
    else:
        return float_n == int_n


movies_df = movies_df[movies_df['id'].apply(is_integer)]
```
